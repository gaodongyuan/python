```python
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
tips_data = sns.load_dataset("tips")



box_obj = tips_data.plot(kind="box",y=['total_bill','tip'],grid=True,fontsize =20,title='total_bill and tip',figsize=(10,5))



box_obj.axes.title.set_size(25)
```

```python
tips_data = sns.load_dataset("tips")



tips_data.plot(kind="scatter",x='total_bill',y='tip',xticks  =[-10,10,30,50,70])



tips_data.groupby('day').sum().plot(kind="barh",y=["total_bill","tip"],stacked =True)



tips_data.groupby("day").count().plot(kind="pie",y="tip",autopct="%0.3f%%",figsize=(6,6))
```
